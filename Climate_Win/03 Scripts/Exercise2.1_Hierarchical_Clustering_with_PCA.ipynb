## 3. Hierarchical clustering

### 3.1 Hierarchical Clustering - Single

```python
# Hierarchical Clustering - single method

dfyear2020_pca_single = linkage(df_pca.loc[:,
['PCA1',
'PCA2',
'PCA3',
'PCA4',
'PCA5',
'PCA6',
'PCA7',
'PCA8',
'PCA9',
'PCA10',
'PCA11',
'PCA12',
'PCA13',
'PCA14',
'PCA15',
'PCA16',
'PCA17',
'PCA18']],method="single")
plt.figure(figsize=(18,6))
dendrogram(dfyear2020_pca_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Hierarchical Clustering with PCA - Single Method",fontsize=18)
plt.show()
```

### 4.2 Hierarchical Clustering - Complete Method

```python
# Hierarchical Clustering - complete method

dfyear2020_pca_complete = linkage(df_pca.loc[:,
['PCA1',
'PCA2',
'PCA3',
'PCA4',
'PCA5',
'PCA6',
'PCA7',
'PCA8',
'PCA9',
'PCA10',
'PCA11',
'PCA12',
'PCA13',
'PCA14',
'PCA15',
'PCA16',
'PCA17',
'PCA18']],method="complete")
plt.figure(figsize=(18,6))
dendrogram(dfyear2020_pca_complete, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Hierarchical Clustering with PCA - Complete Method",fontsize=18)
plt.show()
```

### 4.3 Hierarchical Clustering - Average Method

```python
# Hierarchical Clustering - average method

dfyear2020_pca_avg = linkage(df_pca.loc[:,
['PCA1',
'PCA2',
'PCA3',
'PCA4',
'PCA5',
'PCA6',
'PCA7',
'PCA8',
'PCA9',
'PCA10',
'PCA11',
'PCA12',
'PCA13',
'PCA14',
'PCA15',
'PCA16',
'PCA17',
'PCA18']],method="average")
plt.figure(figsize=(18,6))
dendrogram(dfyear2020_pca_avg, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Hierarchical Clustering with PCA - Average Method",fontsize=18)
plt.show()
```

### 4.4 Hierarchical Clustering - Ward Method

```python
# Hierarchical Clustering - ward method

dfyear2020_pca_ward = linkage(df_pca.loc[:,
['PCA1',
'PCA2',
'PCA3',
'PCA4',
'PCA5',
'PCA6',
'PCA7',
'PCA8',
'PCA9',
'PCA10',
'PCA11',
'PCA12',
'PCA13',
'PCA14',
'PCA15',
'PCA16',
'PCA17',
'PCA18']],method="ward")
plt.figure(figsize=(18,6))
dendrogram(dfyear2020_pca_ward, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Hierarchical Clustering with PCA - Ward Method",fontsize=18)
plt.show()
```
