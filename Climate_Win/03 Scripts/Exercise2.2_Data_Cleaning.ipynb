## 3. Data cleaning

```python
# Dropping date and month column for df_cw
df_cw_nodate = df_cw.drop(['DATE','MONTH'], axis=1)
```

```python
# Check output
df_cw_nodate
```

```python
# Remove all the columns containing the name of three cities: GDANSK, ROMA, TOURS which don't exist in Pleasant weather data set
# Create a pattern 
pattern=r'(GDANSK|ROMA|TOURS)'

# Filter the columns to remove
columns_to_remove=df_cw_nodate.filter(regex=pattern).columns
```

```python
# Check columns_to_remove
columns_to_remove
```

```python
# Remove the columns and create a new data set
df_cw_nodate_1=df_cw_nodate.drop(columns=columns_to_remove, inplace=False)
```

```python
# Check output
df_cw_nodate_1
```

```python
# Check for missing value
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print (df_cw_nodate_1.isnull().sum()) 
```

#### 6 stations with missing windspeed column.

```python
# Remove all the columns containing the two observations: wind_speed and snow_depth  
# Create a pattern 
pattern_1=r'(wind_speed|snow_depth)'

# Filter the columns to remove
columns_to_remove_1=df_cw_nodate_1.filter(regex=pattern_1).columns 
```

```python
# Check columns_to_remove_1
columns_to_remove_1
```

```python
# Remove the columns
df_cw_nodate_2=df_cw_nodate_1.drop(columns=columns_to_remove_1, inplace=False)
```

```python
# Check output
df_cw_nodate_2
```

```python
# Add three new columns cloud_cover, pressure and humidity for three locations and copy the values from the closest weather staions
df_cw_nodate_2[['KASSEL_cloud_cover', 'MUNCHENB_pressue', 'STOCKHOLM_humidity']]=df_cw_nodate_2[['LJUBLJANA_cloud_cover', 'SONNBLICK_pressure', 'OSLO_humidity']].copy()
```

```python
# Check output
df_cw_nodate_2
```

```python
# Re-sort the columns in order
df_cw_sort = df_cw_nodate_2[sorted(df_cw_nodate_2.columns)]
```
