## 4. Scaling

```python
# Create scaler object using StandardScaler from sklearn.preprocessing
# StandardScaler assumes data is normally distributed and scales with a distribution
# Around 0 and standard deviation of 1. Scaling happens independently with each variable.

scaler = StandardScaler()
```

```python
# Convert columns name into a list
columns = df_cw.columns.tolist()
```

```python
# Check output
columns
```

```python
# Create new df with scaled data

df_cw_scaled = pd.DataFrame(scaler.fit_transform(df_cw), columns=columns)
df_cw_scaled.head()
```

```python
# Compare the original vs. scaled data

df_cw.head()
```

```python
# Dropping date and month for scaled data
df_cw_scaled_nodate = df_cw_scaled.drop(['DATE','MONTH'], axis=1)
```

```python
df_cw_scaled_nodate
```

```python
# Combined unscaled date column with scaled data
df_cw_scaled_date = pd.concat([df_cw[['DATE','MONTH']],df_cw_scaled_nodate], axis=1)
```

```python
# Check output
df_cw_scaled_date
```

## 5. Export data
